```python
import cv2
import numpy as np
import pickle

# Define the common target shape as a tuple (width, height)
target_shape = (84, 84)

data_dict = pickle.load(open('./data.pickle', 'rb'))

# Create an empty NumPy array to store the resized images
resized_images = np.empty((len(data_dict['data']), *target_shape), dtype=np.uint8)

# Resize the images and store them in the new array
for i, img in enumerate(data_dict['data']):
    img = np.array(img)
    if img.shape != target_shape:
        img = cv2.resize(img, target_shape)
    resized_images[i] = img

# Convert the resized_images array to a NumPy array
data = np.array(resized_images)

for img in data_dict['data']:
    print(np.array(img).shape)
```
